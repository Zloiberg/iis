```python
"""
Лабораторная работа 4.
Построим  абстрактную фабрику производителей пицц состоящей из
Pizzahut, Dominos, Papajhons, которые будут выпускать пиццы разных начинок.
Также с помощью builder введем дополнительные модификации: доп сыр, доп мясо.
"""
from abc import ABC, abstractmethod, abstractproperty
from enum import Enum
import random

class body(Enum):
    pepperoni = "pepperoni"
    pineapple = "pineapple"
    barbecue = "barbecue"

class AbstractFactory(ABC):
    @abstractmethod
    def create_pepperoni(self, addcheese=None, addmeat=None):
        pass

    @abstractmethod
    def create_pineapple(self, addcheese=None, addmeat=None):
        pass

    @abstractmethod
    def create_barbecue(self, addcheese=None, addmeat=None):
        pass

class Builder(ABC):
    @abstractmethod
    def body(self) -> None:
        pass
    
    @abstractproperty
    def product(self) -> None:
        pass

    @abstractmethod
    def addcheese(self) -> None:
        pass

    @abstractmethod
    def addmeat(self) -> None:
        pass
    
class PizzahutFactory(AbstractFactory):
    def create_pepperoni(self, addcheese=None, addmeat=None):
        b = PizzaBuilder('Pizzahut').body(body.pepperoni).addcheese(addcheese).addmeat(addmeat)
        return b.product() 

    def create_pineapple(self, addcheese=None, addmeat=None):
        b = PizzaBuilder('Pizzahut').body(body.pineapple).addcheese(addcheese).addmeat(addmeat)
        return b.product()

    def create_barbecue(self, addcheese=None, addmeat=None):
        b = PizzaBuilder('Pizzahut').body(body.barbecue).addcheese(addcheese).addmeat(addmeat)
        return b.product()

class DominosFactory(AbstractFactory):
    def create_pepperoni(self, addcheese=None, addmeat=None):
        b = PizzaBuilder('Dominos').body(body.pepperoni).addcheese(addcheese).addmeat(addmeat)
        return b.product() 

    def create_pineapple(self, addcheese=None, addmeat=None):
        b = PizzaBuilder('Dominos').body(body.pineapple).addcheese(addcheese).addmeat(addmeat)
        return b.product() 

    def create_barbecue(self, addcheese=None, addmeat=None):
        b = PizzaBuilder('Dominos').body(body.barbecue).addcheese(addcheese).addmeat(addmeat)
        return b.product() 

class PapajhonesFactory(AbstractFactory):
    def create_pepperoni(self, addcheese=None, addmeat=None):
        b = PizzaBuilder('Papajhones').body(body.pepperoni).addcheese(addcheese).addmeat(addmeat)
        return b.product()   

    def create_pineapple(self, addcheese=None, addmeat=None):
        b = PizzaBuilder('Papajhones').body(body.pineapple).addcheese(addcheese).addmeat(addmeat)
        return b.product()

    def create_barbecue(self, addcheese=None, addmeat=None):
        b = PizzaBuilder('Papajhones').body(body.barbecue).addcheese(addcheese).addmeat(addmeat)
        return b.product()
    
class Pizza():
    def __init__(self, brand):
        self.brand = brand
        self.body = None
        self.addcheese = None
        self.addmeat = None

class PizzaBuilder(Builder):
    def __init__(self, brand):
        self.brand = brand
        self.reset()

    def reset(self):
        self._product = Pizza(self.brand)
        
    def addcheese(self, addcheese=None):
        self._product.addcheese = random.choice(["mozarella", "parmesan", "brie", "no cheese"])
        return self
    
    def addmeat(self, addmeat=None):
        self._product.addmeat = random.choice(["pepperoni", "beef", "chicken", "no meat"])
        return self
    
    def product(self):
        product = self._product
        self.reset()
        return product

    def body(self, body):
        self._product.body = body.name
        return self

def Order(factory : AbstractFactory):
    pepperoni = factory.create_pepperoni(addcheese=random.choice, addmeat=random.choice)
    pineapple = factory.create_pineapple(addcheese=random.choice, addmeat=random.choice)
    barbecue = factory.create_barbecue(addcheese=random.choice, addmeat=random.choice)

    if pepperoni:
        print(f"{pepperoni.brand} {pepperoni.body} pizza with added {pepperoni.addcheese} and {pepperoni.addmeat} ")

    if pineapple:
        print(f"{pineapple.brand} {pineapple.body} pizza with added {pineapple.addcheese} and {pineapple.addmeat} ")

    if barbecue:
        print(f"{barbecue.brand} {barbecue.body} pizza with added {barbecue.addcheese} and {barbecue.addmeat} ")


if __name__ == "__main__":
    Order(DominosFactory())
    Order(PizzahutFactory())
    Order(PapajhonesFactory())
```
